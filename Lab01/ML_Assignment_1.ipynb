```python
# Q2. Write 5 different functions and call them in one program.
```

```python
# Function 1
def add(a,b):
    return a+b
# Function 2
def subtract(a,b):
    return a-b
# Function 3
def prod(a,b):
    return a*b
# Function 4
def divide(a,b):
    return a/b
# Function 5
def power(a,b):
    return a**b
```

```python
# Calculator
choice = input("""Enter the operation you want to perform on the numbers
be it Add, Subtract, Product, Divide, Power: """)
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
if choice == "Add":
    print("The sum is: ",add(num1,num2))
elif choice == "Subtract":
    print("The difference is: ",subtract(num1,num2))
elif choice == "Product":
    print("The product is: ",prod(num1,num2))
elif choice == "Divide":
    print("The division results in: ",divide(num1,num2))
elif choice == "Power":
    print("The resultant is: ",power(num1,num2))
else:
    print("Undefined Operation!")
```
