### Question-2) Write 5 different functions and call them in one program.

```python
def add(a,b):
    c = a+b
    print(f"The sum of {a} and {b} will be: ",c)
    return a+b

def sub(a,b):
    if a>=b:
        c = a-b
        return a-b
    else:
        c = a-b
        return b-a
    print(f"The difference of {a} and {b} is: ",c)
    
def mult(a,b):
    c = a*b
    print(f"The multiplication of {a} and {b} is: ",c)
    return a*b

def div(a,b):
    c = a/b
    print(f"The quotient after dividing of {a} by {b} is: ",c)
    return a/b

def power(a,b):
    c = a**b
    print(f"The power of {a} by {b} is: ",c)
    return a**b
```

```python
print("Welcome to the calculator")
print('''Choose the operation you want to execute:
          MENU
          1 - ADDITION
          2 - SUBTRACTION
          3 - MULTIPLICATION
          4 - DIVISION
          5 - POWER                               ''')
choice = int(input("Enter your choice: "))
num1 = float(input("Enter the number 1: "))
num2 = float(input("Enter the number 2: "))

if choice==1:
    add(num1,num2)
elif choice==2:
    sub(num1,num2)
elif choice==3:
    mult(num1,num2)
elif choice==4:
    div(num1,num2)
elif choice==5:
    power(num1,num2)
else:
    print("You made the wrong choice!!")
```
