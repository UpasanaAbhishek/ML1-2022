### Visualization

```python
sns.relplot(x = "house age", y = "house price of unit area", data = real_estate)
```

```python
sns.relplot(x = "house age", y = "house price of unit area",hue="number of convenience stores", data = real_estate)
```

```python
sns.relplot(x = "house age", y = "house price of unit area",hue="distance to the nearest MRT station", data = real_estate)
```

```python
plt.figure(figsize=(14,6))
sns.scatterplot(x = "longitude", y = "latitude", hue = "number of convenience stores",data = real_estate)
```

```python
sns.relplot(x = "longitude",y = "latitude",hue = "number of convenience stores", style = "house age",height = 6,data = real_estate);
```

```python
sns.relplot(x = "transaction date", y = "house age", data = real_estate)
```

```python
plt.figure(figsize = (12,10))
real_estate.hist()
plt.show()
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(real_estate.corr(), annot=True, cmap = "inferno")
```

```python
sns.countplot(x = "longitude", hue = "number of convenience stores",data = real_estate)
```

```python
plt.figure(figsize = (10,10))
sns.violinplot(x = "transaction date", y = "house price of unit area", data = real_estate)
```

```python
plt.figure(figsize = (20,10))
sns.catplot(x = "transaction date", y = "house price of unit area", data = real_estate)
```
