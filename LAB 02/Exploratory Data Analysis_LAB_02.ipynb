### Heatmap

```python
'''heatmap is a two dimensional visual representation of data where values are encoded in colors,
delivering a convenient, insightful view of information'''

sns.heatmap(correlation, xticklabels=correlation.columns, yticklabels=correlation.columns, 
            annot=True)
```

### Pairplot

```python
'''Pairplot visualizes the given data to find the relationship between them 
where the variables can be continous or categorical'''

sns.pairplot(df)
```

### Catplot(Categorical)

```python
power = pd.read_csv("Power_consumption.csv")
power
```

```python
sns.stripplot(x=power["Temperature"])
```

```python
sns.stripplot(x=power["Wind Speed"])
```

```python
sns.stripplot(x="Wind Speed", y="Humidity", data=power)
```

```python
sns.stripplot(x="Wind Speed", y="Humidity", data=power, jitter=0.05)
```

```python
sns.stripplot(x="Wind Speed", y="Humidity", data=power, jitter=0.05, linewidth = 1)
```

```python
sns.stripplot(x="Temperature", y="Humidity", hue="Wind Speed", data=power, jitter=0.05)
```

```python
sns.stripplot(x="Wind Speed", y="Humidity", hue="diffuse flows",
                   data=power, palette="Set2", size=20, marker="D",
                   edgecolor="gray", alpha=.25)
```

```python
sns.boxplot(x="Wind Speed", y="Humidity", data=power, whis=np.inf)
sns.stripplot(x="Wind Speed", y="Humidity", data=power, color=".3")
```

```python
sns.violinplot(x="Humidity", y="Wind Speed", data=power, inner=None, color=".8")
sns.stripplot(x="Humidity", y="Wind Speed", data=power)
```
