### Visualisations of the Data

## Relational Plots: used for visualizing the statistical relationship between the data points


```python
# Relational plot between sugar content and density of the wine
sns.relplot(x="residual sugar",y="density",data=wine)
```

```python
# Higher the residual sugar content, higher the density
# As the density increases, the sugar content also tends to increase
```

```python
# Add kind argument(=line)
sns.relplot(x="residual sugar",y="density",data=wine,kind="line")
```

```python
# Relational plot between alcohol and density
sns.relplot(x="alcohol",y="quality",data=wine,kind="line")
```

```python
# Higher the alcohol content, higher the quality of the wine
```

### Five extra charts

### 1. Single Line Plot: presents data on x-y axis using a line joining datapoints

```python
# Single line plot between citric acid and volatile acidity
sns.lineplot(x="citric acid", y="volatile acidity", data=wine)
sns.set(style="darkgrid")
  
```

```python
# Significant fluctuation between the citric acid content and the volatile acidity, at citric acid content between 0.8 and 1,
# the volatile acidity almost remains constant
```

### 2. Histograms: groups values into ranges (or bins), and the height of a bar shows how many values fall in that range

```python
#Histogram of the chlorides present in the wine
sns.histplot(x="chlorides", data=wine);
```

```python
# Maximum values for chlorides present in the wine fall between 0 and 0.1
```

```python
# Histogram of the pH level of the wine
sns.histplot(x='pH', data=wine);
```

```python
# Maximum pH values of the wine lie between 3.2 and 3.4
```

###  3. KDE plot: The kernel density estimate plot creates a smooth version of a histogram by normalizing all points to appear under one curve


```python
sns.kdeplot(x="pH", data=wine)
```

```python
# Maximum pH values of the wine are in the range 3.2 and 3.4
```

### 4. Rug plot: draws ticks on the x-axis that show the location of individual data points(complement histograms when it comes to outliers because we can see where the outlier data points fall)

```python
sns.rugplot(x='sulphates', data=wine, height=.03, color='darkblue')
```

```python
# Maximum values for the sulphate content present in the wine, range between 0.25 to 1.25,i.e.,most wines have sulphate 
# content in this range. There are present some extreme values for the sulphate content as well(can be treated as outliers)
```

### 5. Strip plot: implements a scatter plot to show the spread of individual observations for a feature.

```python
sns.stripplot(x=wine["alcohol"]);
```

```python
# There are many overlapping points in the range 9 to 12,i.e., 
#many wines have alcohol content in this range
#and there are some outliers present as well for the alcohol content.
```

### Fit one SLR and one MLR model using sklearn. Check if the assumptions of LR have been met

```python
sns.regplot(x="alcohol", y="quality", data = wine)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
sns.set(style="darkgrid")
```

### Simple Linear Regression

```python
x = wine.alcohol
```

```python
y = wine.quality
```

```python
model = LinearRegression()
```

```python
model.fit(np.array(x).reshape(-1,1),y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
model.score(np.array(x).reshape(-1,1),y)
```

### Multiple Regression


```python
X = wine[['citric acid', 'fixed acidity', 'volatile acidity']]
y = wine['quality']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=4)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
model.score(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
y_test.values
```

```python
model.score(X_test, model.predict(X_test))
```

```python
model.score(X_train, model.predict(X_train))
```

```python
model.score(X_test, y_test)
```

```python
model.score(X_train, y_train)
```
