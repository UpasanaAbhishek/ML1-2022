<h3><u>II. Logistic Regression </u></h3>

```python
#Required Package
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
model = LogisticRegression()
```

```python
#Setting the independent and dependent variables
x = data[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']]
y = data['Outcome']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
#There is a 76.623% accuracy of this model
```

<h3><u>III. Evaluation Metrics </u></h3>

<b>a) Confusion Matrix</b><br>
It is a summary of prediction results on a classification problem.The confusion matrix shows the ways in which your classification model is confused when it makes predictions.<br>
It is a n x n matrix where n is the number of classes.<br>

```python
y_pred = model.predict(x_test)
```

```python
len(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cf_matrix = confusion_matrix(y_pred,y_test)
```

```python
cf_matrix
```

```python
plot9 = sns.heatmap(cf_matrix/np.sum(cf_matrix), annot=True, fmt='.2%', cmap='gnuplot')
plot9.set_xlabel('Predicted Values')
plot9.set_ylabel('Actual Values');
plot9.xaxis.set_ticklabels(['False','True'])
plot9.yaxis.set_ticklabels(['False','True'])

#Out of the 154 predictions, the model predicted True (or 1 or that a person will be diabetic 24% of the time)
#And No - 76% of the time. 
#True Negatives - 57.14%
#False Positives - 4.55%
#False Negatives - 18.83%
#True Positives - 19.48%
```

<b>b) Precision Score</b>

```python
from sklearn.metrics import precision_score, recall_score, f1_score
prec_score = precision_score(y_test, y_pred)
prec_score
#The precision score is 81%. This means that when the model predicts that a person is diabetic
#there is a 81% chance that it is true. 
```

<b>c) Recall Score </b>

```python
r_score = recall_score(y_test, y_pred)
r_score
#The model correctly identifies diabetics ~51% of the time
```

<b>d) F1 Score </b>

```python
f1score = f1_score(y_test, y_pred, labels=None, pos_label=1, average='binary', sample_weight=None, zero_division='warn')
f1score
#Range is between 0 and 1. Closer to 1, better model. 
#Therefore, we have a moderately good model.
```

<b>e) Sensitivity </b>

```python
sensitivity = cf_matrix[0,0]/(cf_matrix[0,0]+cf_matrix[0,1])
print('Sensitivity : ', sensitivity)
#Sensitivity is the proportion of true positives that are correctly predicted by the model 
#~75.21% of the true positives are correctly predicted by our model
```

<b>f) Specificity </b>

```python
specificity = cf_matrix[1,1]/(cf_matrix[1,0]+cf_matrix[1,1])
print('Specificity : ', specificity)
#Specificity is the proportion of true negatives that are correctly predicted by the model.
#~81.1% of the true negatives were correctly predicted. 
```

<b>g) Miscalculation Rate</b>

```python
ms_rate = (cf_matrix[0,1]+cf_matrix[1,0])/len(y_pred)
ms_rate
#Therefore, our model incorrectly predicted 23.37% of the times.
```

<b>h) ROC and AUROC</b>

```python
from sklearn import metrics
#defining metrics
y_pred = model.predict_proba(x_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred)

#creating ROC curve
pp.plot(fpr,tpr)
pp.ylabel('True Positive Rate')
pp.xlabel('False Positive Rate')
pp.show()
#Since the curve hugs the top left corner more, it is a good model for classification. 

#Getting AUC value 
auc = metrics.roc_auc_score(y_test, y_pred)
print("AUC Value:",auc)
#Since value is close to 1, it is a good model for classification. 
```
