```python
%matplotlib inline
import pandas as pd
import numpy as np
import sklearn
import sklearn.datasets
import matplotlib.pyplot as plt
import seaborn as sns;sns.set()
```

```python
from sklearn.datasets import fetch_20newsgroups
```

```python
data = fetch_20newsgroups()
data.target_names
```

```python
text_categories = data.target_names
train = fetch_20newsgroups(subset="train", categories=text_categories)
test = fetch_20newsgroups(subset="test", categories=text_categories)
```

```python
print(train.data[5])
```

Importing necessary packages 

```python
from sklearn.feature_extraction.text import TfidfVectorizer #weighing words 
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline # pumping values from Tfidf to MultinomialNM
```

Creating the model based on Multinomial Naive Bayes 

```python
model = make_pipeline(TfidfVectorizer(), MultinomialNB())
```

 Training the model with the train data 


  

```python
model.fit(train.data,train.target)
```

Creating labels for the test data 

```python
labels = model.predict(test.data)
```

Creating confusinn matrix and heat map

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(test.target,labels)
plt.figure(figsize= (16,10))
sns.heatmap(mat.T,square = True,annot = True ,fmt = 'd',cbar = False,xticklabels = train.target_names,yticklabels = train.target_names)
plt.xlabel('true label')
plt.ylabel('predicted label')
```

```python
 def predict_category(s,train = train , model = model):
        pred = model.predict([s])
        return train.target_names[pred[0]]
```
