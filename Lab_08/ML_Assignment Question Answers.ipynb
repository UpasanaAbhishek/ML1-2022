## 1. Write a program that uses an if-else ladder.

```python
mark = int(input("Enter your score out of 100 : "))
if mark >= 0 and mark <= 100:
    if mark >= 90:
        print("You've got A+")
    elif mark >= 80:
        print("You've got A")
    elif mark >= 70:
        print("You've got B+")
    elif mark >= 60:
        print("You've got B")
    elif mark >= 50:
        print("You've got C+")
    elif mark >= 40:
        print("You've got D")
    else:
        print("You've got F")
else:
    print("Invalid entry")
```

## 2. Write 5 different functions and call them in one program.

```python
def add(a,b):
    c = a + b
    print('The sum of {a} and {b} is = ',c)
def sub(a,b):
    c = a - b
    print('{a} subtracted by {b} is = ',c)
def div(a,b):
    if b==0:
        print('Invalid value of second number')
        return
    c = a/b
    print('{a} divided by {b} is = ',c)
def mul(a,b):
    c = a*b
    print('{a} multiplied by {b} is = ',c)
def power(a,b):
    c = a**b
    print('{a} to the power {b} is = ',c)

print("Choose an option you want to execute \n 1 - Addition \n 2 - Subtraction\n 3 - Division\n 4 - Multiplication\n 5 - Power")

option = int(input('Enter your option : '))
num1 = int(input('Enter the first number : '))
num2 = int(input('Enter the second number : '))

if option >=1 and option <= 5:
    if option == 1:
        add(num1,num2)
    elif option == 2:
        sub(num1,num2)
    elif option == 3:
        div(num1,num2)
    elif option == 4:
        mul(num1,num2)
    elif option == 5:
        power(num1,num2)
else:
    print('Invalid option')
```

## 5. Can we find correlation between two categorical variables? How?

It is possible to find the correlation between two categorical variables.
Tetrachoric correlation is used to calculate the correlation between binary categorical variables. Polychoric correlation is used to calculate the correlation between ordinal categorical variables. And Cramers V is used to calculate the correlation between nominal categorical variables.
