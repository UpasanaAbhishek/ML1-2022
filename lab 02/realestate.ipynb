```python
#Determine the Features & Target Variable
X=df.drop('Y house price of unit area',axis=1)
y=df['Y house price of unit area']
```

```python
#Split the Dataset to Train & Test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
#Train the Model
from sklearn.linear_model import LinearRegression
```

```python
model= LinearRegression()
model.fit(X_train, y_train)
```

```python
#Coeficient Matrix
pd.DataFrame(model.coef_, X.columns, columns=['Coeficient'])
```

```python
#Predict Test Data
y_pred=model.predict(X_test)
pd.DataFrame({'Y_Test': y_test,'Y_Pred':y_pred }).head(5)
```

```python
#Evalutaing the Model
from sklearn import metrics
MAE= metrics.mean_absolute_error(y_test, y_pred)
MSE= metrics.mean_squared_error(y_test, y_pred)
RMSE=np.sqrt(MSE)

pd.DataFrame([MAE, MSE, RMSE], index=['MAE', 'MSE', 'RMSE'], columns=['Metrics'])
```

```python
#Defining dpeendent and independent variables
#X : idependent variable is going to be (Height)
#Y : dependent variable is going to be (Weight)
X = df["X1 transaction date"].values
Y = df["X2 house age"].values
```

```python
#tesing training and splitting data
#with reshaping the data

x_train, x_test, y_train, y_test = train_test_split(X,Y, train_size = 0.8, test_size = 0.2, random_state = 24)

x_train = x_train.reshape(-1, 1)
x_test = x_test.reshape(-1, 1)
```

```python
#Creating the linear regression model
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
#Checking the accuracy of training data
train_accuracy = model.score(x_train, y_train)
train_accuracy

#Checking the accuracy of testing data
test_accuracy = model.score(x_test, y_test)
test_accuracy
```

```python
#Plotting
%matplotlib inline
plt.scatter(X,Y, color = "blue")
plt.xlabel("transcationdate")
plt.ylabel("hoursdate")
plt.title("Height & Weight")
```

```python
#plotting actual vs predicted data
%matplotlib inline
plt.scatter(x_test, y_test, color = "blue")
plt.scatter(x_test ,y_predict, color = "red")
plt.title("Actual vs predicted data")
```
