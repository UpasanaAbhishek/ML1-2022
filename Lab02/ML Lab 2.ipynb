# SLR

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
sns.regplot(x="sqft_living", y="price", data = data)
```

```python
model_data = data[['price', 'sqft_living']]
model_data
```

```python
x=model_data.price
y=model_data.sqft_living
model=LinearRegression()
model.fit(np.array(x).reshape(-1,1),y)
```

```python
model.intercept_ #b0
```

```python
model.coef_ #b1
```

```python
model.score(np.array(x).reshape(-1,1),y)
```

# MLR

```python
X = data[['sqft_living','bedrooms', 'bathrooms', 'floors','view']]
y = data['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=4)
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict(X_test)
```

```python
model.score(X_test, model.predict(X_test))
```

```python
model.score(X_train, model.predict(X_train))
```
