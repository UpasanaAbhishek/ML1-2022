### Write 5 different functions and call them in one program

```python
def add(x,y):
    return x+y
def sub(x,y):
    return x-y
def mult(x,y):
    return x*y
def division(x,y):
    return x%y
def quotient(x,y):
    return x/y
a=int(input("Enter first number: "))
b=int(input("Enter Second number: "))

c=0
print("+ -> To perform Addition")
print("- -> To perform Subtraction")
print("* -> To perform Multiplication")
print("% -> To get Remainder")
print("/ -> To get quotient")
choice=input("Enter your Choice: ")
if choice=="+":
    c=add(a,b)
elif choice=="-":
    c=sub(a,b)
elif choice=="*":
    c=mult(a,b)
elif choice=="%":
    c=division(a,b)
elif choice=="/":
    c=quotient(a,b)
else:
    print("Wrong Choice!!")
print(c)
```

### Bonus:Can we find correlation between two categorical variables? How?

#### Yes, correlation can be found between 2 categorical variables by using the chi-squared test

```python
df.columns
```

```python
CrosstabResult=pd.crosstab(index=df['season'],columns=df['cap-color'])
print(CrosstabResult)
```

```python
#H0 is always the assumption that there is no correlation
# importing the required function
from scipy.stats import chi2_contingency
 
# Performing Chi-sq test
ChiSqResult = chi2_contingency(CrosstabResult)
 
# P-Value is the Probability of H0 being True
# If P-Value > 0.05 then only we Accept the assumption(H0)
 
print('The P-Value of the ChiSq Test is:', ChiSqResult[1])
```
